```python


import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.linear_model import LogisticRegression # to apply the Logistic regression
from sklearn.model_selection import train_test_split # to split the data into two parts
from sklearn.cross_validation import KFold # use for cross validation
from sklearn.model_selection import GridSearchCV# for tuning parameter
from sklearn.ensemble import RandomForestClassifier # for random forest classifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm # for Support Vector Machine
from sklearn.metrics import accuracy_score
from sklearn import metrics # for the check the error and accuracy of the model
import matplotlib.pyplot as plt # this is used for the plot the graph 
import seaborn as sns # used for plot interactive graph. I like it most for plot

movies = pd.read_csv("C:\\Users\\saurabh\\Downloads\\imdb-5000-movie-dataset\\movie_metadata.csv")
movies.info()

movies.fillna(value=0,axis=1,inplace=True)
movies = movies.select_dtypes(include=['int64','float'])

features = ['actor_3_facebook_likes', 'actor_1_facebook_likes', 'gross',
       'num_voted_users', 'cast_total_facebook_likes', 'facenumber_in_poster',
       'num_user_for_reviews', 'budget', 'title_year',
       'actor_2_facebook_likes', 'aspect_ratio',
       'movie_facebook_likes']
target = ['imdb_score']

train, test = train_test_split(movies,test_size=0.20)

train_x = train[features] 
train_y = train[target]
test_x = test[features]
test_y = test[target]

#from sklearn.preprocessing import StandardScaler
#sc = StandardScaler()
#train_x_std = pd.DataFrame(sc.fit_transform(train_x),columns=features)
#test_x_std = pd.DataFrame(sc.transform(test_x), columns=features)

#train_x_std.head(2)

corr = train_x.corr()
#draw the heatmap of correlation to identify any features that can be dropped
plt.figure(figsize=(18,18))
sns.heatmap(corr, cbar = True,  square = True, annot=True, fmt= '.2f',annot_kws={'size': 15},
            cmap= 'YlGnBu') 
sns.despine()
plt.tight_layout()


#Convert target into a list as per the expectation of the algorithm
train_y = np.asarray(train_y, dtype="|S6")
test_y = np.asarray(test_y,dtype="|S6")

from sklearn.linear_model import LogisticRegression
logis = LogisticRegression()
logis.fit(train_x, train_y)
logis_score_train = logis.score(train_x, train_y)
print("Training score: ",logis_score_train)
logis_score_test = logis.score(test_x, test_y)
print("Testing score: ",logis_score_test)
predicted = logis.predict(test_x)
algo1 = accuracy_score(test_y, predicted)
print("Accuracy Score:", algo1)

from sklearn.svm import SVC
svm = SVC()
svm.fit(train_x, train_y)
svm_score_train = svm.score(train_x, train_y)
print("Training score: ",svm_score_train)
svm_score_test = svm.score(test_x, test_y)
print("Testing score: ",svm_score_test)
predicted = svm.predict(test_x)
algo2 = accuracy_score(test_y, predicted)
print("Accuracy Score:", algo2)


from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(train_x, train_y)
knn_score_train = knn.score(train_x, train_y)
print("Training score: ",knn_score_train)
knn_score_test = knn.score(test_x, test_y)
print("Testing score: ",knn_score_test)
predicted = knn.predict(test_x)
algo3 = accuracy_score(test_y, predicted)
print("Accuracy Score:", algo3)

from sklearn import tree
dt = tree.DecisionTreeClassifier()
dt.fit(train_x, train_y)
dt_score_train = dt.score(train_x, train_y)
print("Training score: ",dt_score_train)
dt_score_test = dt.score(test_x, test_y)
print("Testing score: ",dt_score_test)
predicted = dt.predict(test_x)
algo4 = accuracy_score(test_y, predicted)
print("Accuracy Score:", algo4)

from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(train_x, train_y)
rfc_score_train = rfc.score(train_x, train_y)
print("Training score: ",rfc_score_train)
rfc_score_test = rfc.score(test_x, test_y)
print("Testing score: ",rfc_score_test)
predicted = rfc.predict(test_x)
algo5 = accuracy_score(test_y, predicted)
print("Accuracy Score:", algo5)

models = pd.DataFrame({
        'Model'          : ['Logistic Regression', 'SVM', 'kNN', 'Decision Tree', 'Random Forest'],
        'Training_Score' : [logis_score_train, svm_score_train, knn_score_train, dt_score_train, rfc_score_train],
        'Testing_Score'  : [logis_score_test, svm_score_test, knn_score_test, dt_score_test, rfc_score_test],
	'Accuracy_score' : [algo1, algo2, algo3, algo4, algo5]
    })
models.sort_values(by='Testing_Score', ascending=False)

```
